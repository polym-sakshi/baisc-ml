Importing all the required libraries

```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Reading the csv file

```python
df=pd.read_csv('homeprices.csv')
df
```

Plotting the graph against Area v/s price

```python
plt.scatter(df.area,df.price,marker='*')
```

Creating an object for LinearRegression code

```python
model=LinearRegression()
```

Train the model using fit method

```python
model.fit(df[['area']],df.price)
```

This is to check 'm' value

```python
model.coef_
```

This is to check 'b' value

```python
model.intercept_
```

Now we can predict using our model for new values

```python
model.predict([[2025]])
```

```python
plt.plot(df.area,model.predict(df[['area']]),color='red')
plt.scatter(df.area,df.price,marker='+',color='red')
```
